### 7. For each book in goodbooks, compute the list of its authors. Assuming that the number of reviews with a text (column work_text_reviews_count) is split equally among all authors, find for each authors the total number of reviews with a text. We will call this quantity the shared number of reviews with a text.

```python
goodbooks['authors']
```

```python
autori = goodbooks['authors'].str.split(', ', expand=True)
autori
```

```python
work_text_reviews_per_autore = []
for i in range(0, autori.shape[0]):
    conta = 0
    for j in range(0, autori.shape[1]):
        if autori[j][i] != None:
            conta +=1
    risultato = goodbooks['work_text_reviews_count'][i] / conta
    work_text_reviews_per_autore.append(risultato)
autori.insert(5, column = "work_text_reviews_per_autore", value = work_text_reviews_per_autore)
autori
```

```python
autori_list = []
for i in range(0, autori.shape[1] - 1):
    autori_list = list(set().union(autori_list, autori[i]))
autori_list.remove(None)
autori_list.sort()
autori_df = pd.DataFrame(autori_list, columns= ["autore"])
autori_df
```

```python
shared_col = []
for i in range(0, autori_df.shape[0]):
    somma = 0
    for j in range(0, autori.shape[0]):
        for k in range(0,autori.shape[1] - 1):
            if autori_df['autore'][i] == autori[k][j]:
                somma += autori['work_text_reviews_per_autore'][j]
    shared_col.append(somma)
autori_df.insert(1, column = "shared_number_of_reviews_with_a_text", value = shared_col)
autori_df
```

### 8. For each year of publication, determine the author that has the largest value of the shared number of reviews with a text.

```python
goodbooks['original_publication_year'].unique()
```

```python
for indice in range(0, len(goodbooks['original_publication_year'].unique())):
    work_text_reviews_per_autore_per_anno = []
    for a in range(0, autori_df.shape[0]):
        somma = 0
        for i in range(0, autori.shape[0]):
            for j in range(0,autori.shape[1] - 1):
                if autori[j][i] == autori_df['autore'][a]:
                    if goodbooks['original_publication_year'][i] == goodbooks['original_publication_year'].unique()[indice]:
                        somma += autori['work_text_reviews_per_autore'][i]
        work_text_reviews_per_autore_per_anno.append(somma)
    autori_df.insert(indice+2, column = f"anno_{goodbooks['original_publication_year'].unique()[indice]}", value = work_text_reviews_per_autore_per_anno)
autori_df
```

```python
for indice in range(0, len(goodbooks['original_publication_year'].unique())):
    autore = autori_df[autori_df[f"anno_{goodbooks['original_publication_year'].unique()[indice]}"] == autori_df[f"anno_{goodbooks['original_publication_year'].unique()[indice]}"].max()]
    print(f"anno {goodbooks['original_publication_year'].unique()[indice]}: ", end="")
    print(autori_df[f"anno_{goodbooks['original_publication_year'].unique()[indice]}"].max(), end="")
    for i in range(0, autore.shape[0]):
        print(" ", autore.iloc[i]['autore'], end="")
    print()
```

### 13. Find the author with the highest average rating according to the goodbooks datasets.

```python
ratings_autori = dict()
for i, row in goodbooks.iterrows():
    autori = row['authors'].split(',')
    for author in autori:
        author = author.strip()
        if author not in ratings_autori:
            ratings_autori[author] = []
        ratings_autori[author].append(row['average_rating'])
ratings_autori
```

```python
for key, value in ratings_autori.items():
    mean = statistics.mean(value)
    ratings_autori[key] = mean
ratings_autori
```

```python
authors_df = pd.DataFrame(ratings_autori.items())
authors_df.rename({0:'Autore', 1:'Media'}, axis='columns', inplace=True)
authors_df.head()
```
